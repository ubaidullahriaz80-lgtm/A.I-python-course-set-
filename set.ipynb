# set

```python
# create a set
set = {1,2,3,4}
print(set)
```

```python
# union in set
a = {1,5,6}
b = {2,3,4}
print('set a',a)
print('set b',b)
print('union of sets is :',a|b)
```

```python
# intersection of set
c = {1,2,3,4,5}
d = {4,5,6,7,8}
print('set c',c)
print('set d',d)
print('intersection of set c nad d :',c&d)
```

```python
# difference of set
e = {1,2,3,4}
f = {1,2,5,6}
print('set e',e)
print('setb f',f)
print('difference of set e and f :',e-f)
```

```python
# updating a set
companies = {'Lacoste', 'Ralph Lauren'}
tech_companies = ['apple', 'google', 'apple']
print('companies',companies)
print('tech_companies',tech_companies)
companies.update(tech_companies)
print(companies)
```

```python
# adding value in set
numbers = {21, 34, 54, 12}
print('Initial Set:',numbers)
numbers.add(32)
print('Updated Set:', numbers)
```
